```python
import wikipediaapi
import pandas as pd
import os


wiki = wikipediaapi.Wikipedia(
    user_agent="NuclearPowerAssignment/1.0 (yourname@email.com)", 
    language='en'
)

#
topics = ["Manhattan Project", "Los Alamos Laboratory", "Trinity (nuclear test)"]


folder_name = 'nuclear_power_files'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

all_data = []


for topic in topics:
    page = wiki.page(topic)
    if page.exists():
       
        individual_df = pd.DataFrame([{"Source": topic, "Content": page.text}])
        
      
        filename = f"{folder_name}/{topic.replace(' ', '_').lower()}.csv"
        individual_df.to_csv(filename, index=False, encoding='utf-8')
        
       
        all_data.append({"Source": topic, "Content": page.text})
        print(f"Saved individual CSV: {filename}")


master_df = pd.DataFrame(all_data)
master_filename = 'us_nuclear_power_master_dataset.csv'
master_df.to_csv(master_filename, index=False, encoding='utf-8')

print(f"\nSuccess! Master Dataset created: {master_filename}")
```
